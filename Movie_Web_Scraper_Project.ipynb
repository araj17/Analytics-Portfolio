```python
table_data = data.find('table',class_='wikitable')
rows = table_data.find_all('tr')
```

```python
rows
```

```python
movies = []
for row in rows:
    curr_movie = []
    for cell in row.find_all("td"):
        if cell.string != "\n" and cell.string != None:
            curr_movie.append(cell.string)
        else:
            curr_movie.append("NA")
    movies.append(curr_movie)
    #print(curr_movie)
print(movies)
```

```python
import pandas as pd

df = pd.DataFrame()

for row in movies:
    if len(row) != 0:
        df = df.append({'Movie Name' : row[0],
                       'Director' : row[1],
                       'Cast' : row[2],
                       'Genre': row[3]},ignore_index= True)
```
